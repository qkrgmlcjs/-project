```python
from urllib.request import urlopen
```

```python
f = urlopen('http://hanbit.co.kr/store/books/full_book_list.html')
```

```python
encoding = f.info().get_content_charset(failobj="utf-8") #HTTP헤더를 기반으로 인코딩 방식을 추출(명시도ㅔ 있지 않을 경우 utf-8)
```

```python
print('encoding :',encoding)
```

```python
text = f.read().decode(encoding)#추출한 인코딩 방식으로 디코딩
```

```python
import re
import sys
from urllib.request import urlopen
```

```python
f = urlopen('http://hanbit.co.kr/store/books/full_book_list.html')
```

```python
bytes_content = f.read()
```

```python
scanned_text = bytes_content[:1024].decode('ascii',errors='replace')
```

```python
match = re.search(r'charset=["\']?([\w-]+)',scanned_text)
```

```python
match.group(1)
```

```python
if match: 
    encoding = match.group(1)
else : 
    encoding = 'utf-8'
```
