# 머신러닝
### 예측 그런거

- 인간이 정확히 하나하나 로직을 지정해주기 어려운 복잡한 문제를 데이터에 기반한 학습을 통해 해결

- 머신러닝 알고리즘을 사용할때 가장 중요한 부분은 머신러닝 모델이 잘 학습할 수 있도록 적절한 특징을 설정해주는것

- 청바지판매 -> 광고를 위한 조건 남자인지 여자인지, 나이, 소득수준 등

- 예측 모델을 갖고 있을 경우, 중요한 비즈니스적 의사결정을 안정적이고 계획적으로 수행할 수 있습니다.

# 지도학습(supervised Learning)

- 지도학습 방법론을 사용하기 위해서는 트레이닝 데이터의 구성이 (인풋 데이터, 데이터에 대한 정답) 쌍으로 구성되어 있어야만 함 즉 지도학습은 정답을 보여주면서 학습시키는 머신러닝

- 보통 인풋 데이터를 x, 데이터에 대한 정답을 y라고 부릅니다.

- 데이터는 (x,y)쌍으로 구성

- ex) 키를 기반으로 몸무게를 예측하는 모델을 만드는 경우 트레이닝 데이터는 여러사람에게서 수집한 키와 몸무게데이터

## Training Data, Test Data
- (x,y)로 구성된 데이터를 트레이닝 데이터 머신러닝 모델을 사용하는 경우 다음 2가지 과정을 거친다

 1. 학습용 데이터로 머신러닝 모델 학습
 2. 트레이닝 데이터에서 따로 빼놓은 테스트 데이터->측정용
 - 트레이닝 테스트 나누기  80:20
 
 # 검증용 데이터(Validation Data)
 - 트레이닝, 검증용, 테스트로 나눈다
 - 검증용 데이터는 트레이닝 과정에서 학습에 사용하지는 않지만 중간중간 테스트하는데 사용해서 모델이 오버피팅에 빠지지않았나 체크 ->정규화의 한가지 방법
 - 트레이닝 데이터의 일부를 검증용으로
 
 ### 오버피팅
- 과적합, 과대적합? -> 트레이닝 데이터에만 너무 최적화 되버림 ->모델이 너무 강력할때
### 언더피팅 
- 트레이닝 중간에서 트레이닝 데이터 조차도 학습이 잘안됨 ->약할때

 ### 분류 문제 : 예측하는 결과값이 이산값,
 ### 회귀 문제 : 예측하는 결과값이 연속값
 
 
 

# 선형회귀 (Linear Regression)
- 선형함수를 이용해서 회귀를 수행 (y = Wx+b)

- x,y는 저희가 가지고 있는 데이터 , w와 b는 데이터에 적합한 값으로 학습될 수 있는 파라미터

# 손실 함수(Loss Function) - MSE
- 머신러닝모델을 학습시키기 위해서는 적절한 파라미터값을 알아내기 위해서 현재 파라미터값이 우리가 풀고자 하는 목적에 적합한 값인지를 측정할 수 있어야 함
- 이를 위해 손실함수를 정의함
- 가장 대표적인것중 하나는 평균제곱오차이다.
